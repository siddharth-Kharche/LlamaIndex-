```python
import os
from dotenv import load_dotenv
from llama_index.core import VectorStoreIndex, Document, Settings
from llama_index.llms.groq import Groq
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
if not groq_api_key:
    raise ValueError("Please set GROQ_API_KEY in your .env file")

# Initialize Groq LLM
llm = Groq(model="llama-3.3-70b-versatile", temperature=0.7, api_key=groq_api_key)

# Initialize HuggingFace Embedding
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5", device="cpu")

Settings.llm = llm
Settings.embed_model = embed_model

print("✅ LLM and Embedding model initialized!")
```

## Step 1: Setup Qdrant

```python
import os
from dotenv import load_dotenv
from qdrant_client import QdrantClient
from llama_index.vector_stores.qdrant import QdrantVectorStore
from llama_index.core import StorageContext

# Load environment variables
load_dotenv()

# Get credentials from .env
url = os.getenv("QDRANT_URL")
api_key = os.getenv("QDRANT_API_KEY")

# Setup Qdrant Cloud
client = QdrantClient(url=url, api_key=api_key)

vector_store = QdrantVectorStore(
    client=client,
    collection_name="rag_tutorial"
)

storage_context = StorageContext.from_defaults(
    vector_store=vector_store
)

print("✅ Qdrant vector store setup!")
```

## Step 2: Create Index and Query

```python
documents = [
    Document(text="Qdrant is a vector similarity search engine."),
    Document(text="It supports filtering and payload queries."),
]

index = VectorStoreIndex.from_documents(documents, storage_context=storage_context)
query_engine = index.as_query_engine()
response = query_engine.query("What is Qdrant?")
print(response)
```
